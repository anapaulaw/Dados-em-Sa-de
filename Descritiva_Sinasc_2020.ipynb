# **COMPLETUDE DA BASE**

---



```python
valores_df = df.count() # aqui chamamos a função count que retorna os valores preenchidos de cada variável presente no dataframe
nulos_df = df.isna().sum() # aqui chamamos a função isna que retorna a quantidades de valores ausentes por variável

resultados1 = pd.concat([valores_df , nulos_df], axis=1)  #com a função concat jutamos os valores obtidos com as consultas acima em um unico resultado
resultados1.columns = ['Counts', 'Missings'] # chamamos os resultados obtidos acima para cada variável presente no dataframe

resultados1  #mostramos os resultados obtidos
```

```python
fig, ax = plt.subplots(figsize=(10, 15)) #defino o tamanho do gráfico

#a função plot foi utilizada para cintagem de valores obtidos no resultados1 ou seja, o eixo y que contem cada variavel contem os counts e missings juntos a função kind escolhemos o tipo do grafico
resultados1.sort_values(by='Counts', ascending=True).plot(
    y=['Counts', 'Missings'], kind='barh', stacked=True, ax=ax, color=['green', 'red']
)

ax.set_xlabel('Contagem') #legenda do eixo x
ax.set_ylabel('Variáveis') #legenda do eixo y
ax.set_title('Proporção de Valores e Valores Ausentes por Variável') #titulo do grafico

plt.show() #mostra o grafico
```

## INFERINDO O TIPO DE DADOS PARA OTIMIZAÇÃO DE MÉMORIA

---



```python
vazias = list(resultados1[resultados1['Counts'] == 0].index)
nao_vazias = [col for col in list(df.columns) if col not in vazias]
```

```python
print('Variáveis sem preenchimento:')
resultados1.T.filter(items=vazias).T
```

```python
df = df[nao_vazias]
```

```python
datas = [col for col in df.columns if (('DT' in col) or ('DT_' in col))]
for data in datas:
    try:
        df[data] = pd.to_datetime(df[data])
    except:
        print('*** Coluna de data',data,'apresenta 1 ou mais registros inconsistentes com o padrão datetime.')
        print()
        df[data] = df[data].astype('object')
```

```python
categoricas = [col for col in df.select_dtypes(exclude=['datetime']).columns if 1 <= df[col].nunique() <= 40]
df[categoricas] = df[categoricas].astype('category')
```

```python
df = df.astype(dict(df.select_dtypes(include=['int64']).apply(pd.to_numeric,downcast='integer').dtypes))
df = df.astype(dict(df.select_dtypes(include=['float64']).apply(pd.to_numeric,downcast='float').dtypes))
```

# **MEDIDAS ESTATÍSTICAS**











```python
medidas = {} #cria um dicionario para armazenas ar medidas descritivas

medidas['Valor Mínimo'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).min() #calcula o valor minimo de cada variavel

medidas['Valor Máximo'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).max() #calcula o valor maximo de cada variavel

medidas['Amplitude'] = df.select_dtypes(exclude=['object','category' , 'datetime64[ns]']).max() - df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).min() #calcula a amplitude de cada variavel

medidas['Média'] = df.select_dtypes(exclude=['object','category' , 'datetime64[ns]']).mean() #calcula o valor da média de cada variavel

medidas['Desvio Padrão'] = df.select_dtypes(exclude=['object','category' , 'datetime64[ns]']).std() #calcula o desvio padrão de cada variavel

medidas['Coeficiente de Variação'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).std() / df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).mean() #calcula o coeficiente de variacao de cada variavel

medidas['Variância'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).var() #calcula a variancia de cada variavel

medidas['1º Quartil'] = df.select_dtypes(exclude=['object','category','datetime64[ns]']).quantile(0.25) #calcula os quartis

medidas['2º Quartil'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).quantile(0.50) #calcula os quartis

medidas['3º Quartil'] = df.select_dtypes(exclude=['object','category', 'datetime64[ns]']).quantile(0.75) #calcula os quartis
```

```python
estatisticas = pd.DataFrame(medidas) # cria um data frame contendo a medidas descritivas calculadas
estatisticas #mostra as medidas calculadas
```

GRÁFICO DE SETORES




```python
for coluna in df.columns: # loop para percorrer  cada coluna do dataframe
    if df[coluna].nunique() == 2:  # condição se cada coluna do dataframe tiver valor unico igual a 2

        dados_agrupados = df[coluna].value_counts(normalize=True, dropna=False) #conta os valores preenchidos de cada coluna e a função dropna é utilizada para que os valores nulos saiam na contegem tambem
        dados_agrupados.plot.pie(autopct='%1.1f%%') # após obter os dados agrupados chamamos a função plot.pie para plotar um grafico de setores mostrando os valores em porcentagem


        plt.title(coluna) #titulo de cada gráfico
        plt.show() #visualizando gráfico
```

 HISTOGRAMAS

---








```python
tipos_de_dados = ['int16', 'int32', 'int64','float16','float32','float64'] # cria uma lista contendo os tipos de dados
```

```python
cols = df.select_dtypes(include=tipos_de_dados).columns  # selecionando as colunas que incluem os tipos de dados que queremos plotar

num_cols = len(cols)  # a função len calcula o número de colunas
num_graficos_por_linha = 2  # defino o número de gráficos por linha que quero exibir
num_linhas = (num_cols + num_graficos_por_linha - 1) // num_graficos_por_linha  # defino o número de linhas

fig, eixos = plt.subplots(num_linhas, num_graficos_por_linha, figsize=(15, 5 * num_linhas))  # tamanho dos gráficos por linha
fig.subplots_adjust(hspace=0.5)

for i, col in enumerate(cols):
    linha, col_idx = divmod(i, num_graficos_por_linha)
    ax = eixos[linha, col_idx]

    ax.hist(df[col], bins=50)  # tamanho da barra de frequência
    ax.set_title(col, fontsize=10)  # defino o tamanho do texto do título dos gráficos

for i in range(num_cols, num_linhas * num_graficos_por_linha):
    fig.delaxes(eixos.flatten()[i])

plt.show()  # mostra os gráficos
```

GRÁFICOS DE BARRAS

---



```python
coluna_vertical = [col for col in categoricas if 1 <= df[col].nunique() <= 12] #classificando as variaveis para plotar o grafico vertircal
coluna_horizontal = [col for col in categoricas if 1 <= df[col].nunique() > 12]
```

```python
for col in coluna_vertical: #loop para cada coluna armazenada em coluna_vertical
    plt.figure(figsize=(8, 4)) #tamanho do grafico
    ax = df[col].value_counts(dropna = False).plot(kind='bar') #a função value_counts conta a frequencia de cada categoria de cada coluna do df , a função sort_values apresenta os valores de ordem decrescente
    ax.set_title(col)  #definindo o titulo do gráfico que sera o nome de cada coluna
    ax.set_xlabel('Valores') #nome dos valores do eixo X
    ax.set_ylabel('Frequência') #nome dos valores do eixo Y
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x()+0.15, p.get_height()+0.5)) #funções para costumzação do testo e valores em exibição no gráfico
    plt.show() #mostra o grafico
```

```python
for col in coluna_horizontal: #loop para cada coluna armazenada em coluna_horizontal

    plt.figure(figsize=(8, 8)) #tamanho do gráfico
    ax = df[col].value_counts(dropna = False).head(30).sort_values().plot(kind='barh' ) #a função value_counts conta a frequencia de cada categoria de cada coluna do df , a função sort_values apresenta os valores de ordem decrescente
    ax.set_title(col) #definindo o titulo do gráfico que sera o nome de cada coluna
    ax.set_xlabel('Valores') #nome dos valores do eixo X
    ax.set_ylabel('Frequência') #nome dos valores do eixo Y
    for p in ax.patches:
        ax.annotate(str(p.get_width()), (p.get_width()+1, p.get_y()+0.5)) #funções para costumzação do testo e valores em exibição no gráfico
    plt.show() #mostra o grafico
```
